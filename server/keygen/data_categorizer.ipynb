```python
adjectives = []
with open("adjectives.txt", "r") as file:
    for line in file:
        adjectives.append(line[:-1])
        
nouns = []
with open('nouns.txt', 'r') as file:
    for line in file:
        if line.strip():
            nouns.append(line[:-1])
```

```python
from openai import OpenAI
client = OpenAI(api_key = "")

def categories_verb(adjective):
    completion = client.chat.completions.create(
        model="gpt-4o-mini",
        temperature=0,
        messages=[
            {"role": "system", "content": 
                '''
    
    List of adjective categories
    0. Obscure/Non-adjectives (e.g. zyzys, hello, ylike)
    
    1. Determiner/Quantity(e.g., two, several, many, few, one)

    2. Opinion/Quality (e.g., beautiful, ugly, strange, lovely)

    3. Size (e.g., big, small, tiny, enormous)

    4. Age (e.g., old, young, new, ancient)

    5. Shape (e.g., round, square, long, flat)

    6. Color (e.g., red, blue, green, dark, bright)

    7. Origin/Nationality (e.g., French, American, Asian, European)

    8. Material (e.g., wooden, metal, plastic, cotton)

    9. Purpose (or Qualifier) (e.g., sleeping [bag], running [shoes], cooking [pot])

    10. Noun (acting as a modifier) (e.g.,**chicken** soup, **wool** sweater, **stone** wall)

    You will be provided an adjective. This adjective will be used in code/key a human will need to remember. Respond with only the number corresponding to the adjectives category.
                '''
            },
            {
                "role": "user",
                "content": adjective
            }
        ]
    )
    print(completion.choices[0].message.content, adjective)
    return completion.choices[0].message.content


def determine_noun(noun):
    completion = client.chat.completions.create(
        model="gpt-4o-mini",
        temperature=0,
        messages=[
            {"role": "system", "content": 
                '''
    List of noun categories
    0. Obscure/Non-noun/Strange spelling (e.g. zyzys, hello, ylike)
    1. Normal/Noun
    You will be provided a noun. This noun will be used in code/key a human will need to remember. Respond with only the number corresponding to the nouns category.
                '''
            },
            {
                "role": "user",
                "content": noun
            }
        ]
    )
    print(completion.choices[0].message.content, noun)
    return completion.choices[0].message.content
```

```python
categorized = [categories_verb(adjective) for adjective in adjectives]
all_categories = [[],[],[],[],[],[],[],[],[],[], []]
for adjective, category in zip(adjectives, categorized):
    try:
        all_categories[int(category)].append(adjective)
    except:
        all_categories[0].append(adjective)
        
determined = [determine_noun(noun) for noun in nouns]
all_determined = [[],[]]
for noun, category in zip(nouns, determined):
    try:
        all_determined[int(category)].append(noun)
    except:
        all_determined[0].append(noun)
        
```

```python
import random

def random_thing(all_categories):
    indexes = two_random()
    return get_random_index(all_categories[indexes[0]]) + " " + get_random_index(all_categories[indexes[1]]) + " " + get_random_index(all_determined[1])

def two_random():
    # Generate three unique random numbers between 1 and 10
    random_numbers = random.sample(range(1, 11), 2)

    # Sort the random numbers
    return sorted(random_numbers)


def get_random_index(strings):
    if not strings:
        raise ValueError("The list is empty")
    return strings[random.randint(0, len(strings) - 1)]

for i in range(0,100):
    print(random_thing(all_categories))
```

```python
with open("adjectives_categorized.txt", "w") as file:
    for category in all_categories[1:]:
        for adj in category:
            file.write(adj + "\n")
        
with open("nouns_categorized.txt", "w") as file:
    for noun in all_determined[1]:
        file.writelines(noun + "\n")
```
